```python
#GET ALL VIDEOS OF A YOUTUBE CHANNEL
```

```python
def get_channel_videos(channel_id):
    
    res = youtube.channels().list(id=channel_id, 
                                  part='contentDetails').execute()
    playlist_id = res['items'][0]['contentDetails']['relatedPlaylists']['uploads']
    
    videos = []
    next_page_token = None
    
    while 1:
        res = youtube.playlistItems().list(playlistId=playlist_id, 
                                           part='snippet', 
                                           maxResults=50,
                                           pageToken=next_page_token).execute()
        videos += res['items']
        next_page_token = res.get('nextPageToken')
        
        if next_page_token is None:
            break
    
    return videos
```

```python
videos = get_channel_videos('UCj22tfcQrWG7EMEKS0qLeEg')
```

```python
for video in videos:
    print(video['snippet']['title'])
```

```python
#GET MOST DISLIKED VIDEOS OF THE CHANNEL
```

```python
def get_videos_stats(video_ids):
    stats = []
    for i in range(0, len(video_ids), 50):
        res = youtube.videos().list(id=','.join(video_ids[i:i+50]),
                                   part='statistics').execute()
        stats += res['items']
        
    return stats
```

```python
video_ids = list(map(lambda x:x['snippet']['resourceId']['videoId'], videos))
```

```python
stats = get_videos_stats(video_ids)
```

```python
len(stats)
```

```python
most_disliked = sorted(stats, key=lambda x:int(x['statistics']['dislikeCount']), reverse=True)
```

```python
for video in most_disliked:
    print(video['id'], video['statistics']['dislikeCount'])
```
